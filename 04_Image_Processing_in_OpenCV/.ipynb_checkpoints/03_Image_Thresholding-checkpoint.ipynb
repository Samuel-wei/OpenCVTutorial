# 1. 简单阈值处理
阈值处理的情况比较简单，如果一个像素值大于某一个阈值，就指定为某个值（例如白色），否则指定另外一个值（例如黑色）。**cv2.threshold()**用于实现这一功能。第一个参数是原始图像，该图像必须是**灰度图像**，第二个参数是用于分类像素的阈值，第三个参数是大于阈值的像素要指定的数值**maxVal**。OpenCV通过第四个参数来确定阈值处理方式，该参数可选的值包括：
- cv2.THRESH_BINARY
- cv2.THRESH_BINARY_INV
- cv2.THRESH_TRUNC
- cv2.THRESH_TOZERO
- cv2.THRESH_TOZERO_INV

在OpenCV的文档中对各种阈值处理的方式进行了清楚的解释，请阅读文档。

该函数会返回两个返回值，第一个返回值**retval**稍后会解释，第二个返回值是**阈值图像**。

```python
import cv2

img = cv2.imread('../data/gradient.png', cv2.IMREAD_GRAYSCALE)
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

> **注意**  
> 要绘制多张图像，需要使用**plt.subplot()**函数，具体用法请查询Matplotlib文档。

# 2. 自适应阈值
在前面的例子中，对整幅图像使用了相同的阈值。但是当图像的不同区域光照条件不同的时候，简单的阈值处理无法得到令人满意的结果。在这情况下，需要使用自适应阈值。该算法计算像素的一个小的邻域的阈值，因此在图像不同的位置会计算得到不同的阈值，这对于存在光照变化的图像会得到更好的结果。

该函数包括三个**特定**的输入和一个输出：
- **Adaptive Method** - 决定阈值如何计算：
    - **cv2.ADAPTIVE_THRESH_MEAN_C** - 阈值是邻域的均值。
    - **cv2.ADAPTIVE_THRESH_GAUSSIAN_C** - 阈值邻域的高斯权重均值。
- **Block Size** - 决定邻域的大小
- **C** - 是一个常量，阈值等于均值减去该常量值。

下面的例子比较了在光照变化图像上，全局阈值处理和自适应阈值处理的结果差异：

```python
import cv2

img = cv2.imread('../data/dave.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.medianBlur(img, 5)

ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

titles = ['Original Image', 'Global Thresholding (v = 127)', 
          'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```

# 3. Otus阈值
在第一例子中，**cv2.threshold()**函数的返回值**retVal**，该返回值是用于Otus二值化的。

在全局阈值中，任意选择的阈值是否是正确的选择，难以确定。只能靠实验来确定。但是考虑一个双模图像（简单的说，双模图像就是该图像的直方图存在两个峰值），可以用两个峰值中间的值作为阈值，这就是Otus二值化算法。简单地说，它利用二值图像的直方图来计算阈值（对于非双模图像，该方法不够准确）。

可以在**cv2.threshold()**函数中使用**cv2.THRESH_OTSU**标志位来使用Otus二值化算法。在使用Otsu算法时，阈值参数直接设置为0。算法将会寻找最优的阈值并返回该值。如果没有使用Otus阈值，**retVal**返回阈值参数的值。

下面的例子中，输入图像是一副噪声图像。在第一种处理中，全局阈值设置为127；第二种处理中，直接应用Otsu阈值；第三种处理中，首先用5x5高斯核对图像进行滤波去除噪声，然后再应用Otsu阈值算法。可以看到噪声滤波对结果有改善。

```python
import cv2

img = cv2.imread('../data/noise.png', cv2.IMREAD_GRAYSCALE)

# 全局阈值
ret1, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Otsu阈值
ret2, th2 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 高斯滤波后再用Otsu阈值
blur = cv2.GaussianBlur(img, (5, 5), 0)
ret3, th3 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 绘制所有图像及直方图
images = [img, 0, th1,
          img, 0, th2,
          blur, 0, th3]
titles = ['Original Noisy Image', 'Histogram', 'Global Thresholding (v = 127)',
          'Original Noisy Image', 'Histogram', 'Otsu\'s Thresholding',
          'Gaussian filtered Image', 'Histogram', 'Otsu\'s Thresholding']

for i in range(3):
    plt.subplot(3, 3, i * 3 + 1), plt.imshow(images[i * 3], 'gray')
    plt.title(titles[i * 3]), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3, 3, i * 3 + 2), plt.hist(images[i * 3].ravel(), 256)
    plt.title(titles[i * 3 + 1]), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3, 3, i * 3 + 3), plt.imshow(images[i * 3 + 2], 'gray')
    plt.title(titles[i * 3 + 2]), plt.xticks([]), plt.yticks([])
```

# 4. Otsu二值化是如何工作的
下面的例子将介绍如何用Python来实现Otsu二值化算法。

由于面向的是双模图像，Otsu算法需要找到一个阈值**(t)**，来最小化**带权重的类内方差**，公式如下：

$$\sigma^2_w(t) = q_1(t)\sigma^2_1(t) + q_2(t)\sigma^2_2(t)$$
其中：
$$q_1(t)=\sum^t_{i=1}P(i) \quad \& \quad q_2(t)=\sum^I_{i=t+1}P(i)$$
$$\mu_1(t) = \sum_{i=1}^t \frac {iP(i)} {q_1(t)} \quad \& \quad \mu_2(t) = \sum_{i=t+1}^I \frac {iP(i)} {q_2(t)}$$
$$\sigma_1^2 = \sum_{i=1}^t[i - \mu_1(t)]^2 \frac {P(i)} {q_1(t)} \quad \& \quad \sigma_2^2 = \sum_{i=1}^t[i - \mu_2(t)]^2 \frac {P(i)} {q_2(t)}$$

实际上就是寻找两个峰值之间的某个数值，使得两个分类的方差最小，可以用下面的Python代码实现：

```python
img = cv2.imread('../data/noise.png', cv2.IMREAD_GRAYSCALE)
blur = cv2.GaussianBlur(img, (5, 5), 0)

# 需要计算归一化的直方图和累积分布函数
hist = cv2.calcHist([blur], [0], None, [256], [0, 256])
hist_norm = hist.ravel() / hist.max()
Q = hist_norm.cumsum()

bins = np.arange(256)

fn_min = np.inf
thresh = -1

for i in range(1, 256):
    p1, p2 = np.hsplit(hist_norm, [i]) # 两个类的分布
    q1, q2 = Q[i], Q[255] - Q[i] # 各个类的累积值
    b1, b2 = np.hsplit(bins, [i]) # 权重
    
    # 计算均值和方差
    esp = 1e-8
    m1 = np.sum(p1 * b1) / (q1 - esp)
    m2 = np.sum(p2 * b2) / (q2 - esp)
    v1 = np.sum(((b1 - m1) ** 2) * p1) / (q1 - esp)
    v2 = np.sum(((b2 - m2) ** 2) * p2) / (q2 - esp)
    
    # 保留令函数最小的i
    fn = v1 * q1 + v2 * q2
    if fn < fn_min:
        fn_min = fn
        thresh = i

# 用OpenCV的函数寻找otsu阈值
ret, otsu = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print(thresh, ret)
```
