# 目标
- 使用OpenCV计算图像的傅里叶变换。

# 1. 理论
傅里叶变换被用于分析不同滤波器的频域特征。对于图像，**2D离散傅里叶变换（DFT）**被用于计算频域。用于计算DFT的快速算法被称为**快速傅里叶变换（FFT）**。关于它的细节可以在任何图像处理或信号处理课本中找到。

对于一个正弦信号，$x(t)=Asin(2\pi ft)$，$f$是信号的频率。通过对信号进行采样来得到的离散信号，有着相同的频域，但是周期范围是$[-\pi, \pi]$或$[0, 2 \pi]$（对于N点DFT是$[0,N]$）。可以把图像看成沿两个方向的采样，同时对X和Y方向进行傅里叶变换就可以得到图像的频率表示。

从直觉上说，若正弦信号的幅值在短时间内变化迅速，我们称之为高频信号。若变化缓慢，则称为低频信号。将这个概念扩展到图像上，图像上幅值变化剧烈的位置对应着边缘点或噪声，因此边缘和噪声是图像的高频成分。如果相应位置幅值变化缓慢，则是低频成分。

# 2. Numpy中的傅里叶变换
首先来看Numpy中的傅里叶变换，Numpy中有一个**FFT**包来实现傅里叶变换。**np.fft.fft2()**提供了返回复数数组的频率变换。该函数需要两个参数，第一个参数是输入的灰度图像，第二个参数是输出数组的大小。如果第二个参数大于输入图像的尺寸，那么输入图像要做padding，如果小于输入图像的尺寸则输入图像要进行裁剪。如果忽略第二个参数，输出数组的尺寸与输入图像相同。

在输出数组中，零频成分（DC成分）位于左上角，如果要移到中间，需要两个方向都移动$\frac N2$。可以用函数**np.fft.fftshift()**完成（移位更易于后续的操作）。计算完频域变换后，就可以计算幅值谱。

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = np.log(np.abs(fshift))

plt.subplot(121), plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
```

上面的结果可以看到，在中间区域亮度较高说明低频成分更多。

现在可以对前面求出的频域信息进行操作，例如进行低通滤波并重建图像（逆傅里叶变换）。例如把频域中央60x60的矩形窗口置0。然后应用**np.fft.ifftshift()**将DC成分移回到左上角。然后用**np.fft.ifft2()**进行逆变换，该函数的结果是复数类型，可以取其绝对值作为结果。

```python
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2
fshift[crow - 30:crow + 30, ccol - 30: ccol + 30] = 0
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

plt.imshow(img_back, 'gray')
```

```python
plt.figure(1, (rows // 40, cols // 40))
plt.imshow(img_back)
plt.annotate('', xy=(400, 200), xytext=(450, 150),
             arrowprops=dict(facecolor='red'))
```

就像在图像梯度教程中看到的，上图表示了高通滤波等价于边缘检测器。同时说明了大部分的图像数据可以用频谱的低频区域表示。前面讲解了在Numpy中计算DFT和IDFT的方法，现在看一下在OpenCV中如何去做。

如果靠近观察最后的结果，尤其是用JET配色的图中可以看到一些失真（例如用红色箭头标注的位置）。该区域的外观存在类似波纹的结构，被称为**涟漪效应**。这是因为使用矩形框进行了频域掩码运算。该掩码被转为正弦波后产生这个问题。因此矩形窗口不被用于滤波，更好的选择是高斯窗口。

# 3. OpenCV中的傅里叶变换
OpenCV中提供了**cv2.dft()**和**cv2.idft()**函数用于傅里叶变换。与Numpy中对应的函数产生的结果相同，但是包含了两个通道。第一个通道是结果的实部，第二个通道是结果的虚部。输入图像首先需要转换为**np.float32**类型。

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)

dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
```

> **注意**  
> 也可以使用**cv2.cartToPolar()**一次性返回频谱强度和相位角。

下面进行逆傅里叶变换，上面的例子创建了一个高通滤波器，这一次将移除图像的高频成分，也就是低通滤波器用于模糊图像。为了实现这一点，首先要创建一个Mask，在低频区域设置为1以通过低频成分，高频区域设置为0以移除高频成分。

```python
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2

# 首先创建掩码，中央区域为1，其他区域为0
mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow-30:crow+30, ccol-30:ccol+30] = 1

# 应用掩码和逆傅里叶变换
fshift = dft_shift * mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])

plt.imshow(img_back, 'gray')
```

> **注意**  
> 通常，OpenCV的函数**cv2.dft()**和**cv2.idft()**要比Numpy对应的函数更快，但是Numpy函数接口更为友好。更多性能的话题，请阅读下面的小节。

# 4. DFT的性能优化
DFT运算对于某些数组的大小会更快，对于数组尺寸是2的幂时最快，对于尺寸为2、3、5的乘积时效率也很高。当考虑提升代码性能的时候，可以在进行离散傅里叶变之前，换改变数组的大小到最优的尺寸（通过填充0）。在OpenCV中需要手工进行0填充。但是对于Numpy，可以指定FFT运算的尺寸并自动进行0填充。

在OpenCV中，提供了**cv2.getOptimalDFTSize()**函数来计算最优的尺寸。该函数的结果可应用于**cv2.dft()**和**np.fft.fft2()**。下面用IPython的Magic命令来衡量其运行时间：

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)
rows, cols = img.shape
print(rows, cols)
```

```python
nrows = cv2.getOptimalDFTSize(rows)
ncols = cv2.getOptimalDFTSize(cols)
print(nrows, ncols)
```

可以看到图像大小$(342, 548)$修改为$(360, 576)$。现在为其进行0填充（对于OpenCV），然后计算DFT性能。可以创建一个大的值为0的数组，然后把图像拷贝进去，或者使用**cv2.copyMakeBorder()**。

```python
nimg = np.zeros((nrows, ncols))
nimg[:rows, :cols] = img
plt.imshow(nimg, 'gray')
```

或者用下面的代码：

```python
right = ncols - cols
bottom = nrows - rows
bordertype = cv2.BORDER_CONSTANT
nimg = cv2.copyMakeBorder(img, 0, bottom, 0, right, bordertype, value=0)
plt.imshow(nimg, 'gray')
```

下面进行Numpy函数的性能比较：

```python
%timeit fft1 = np.fft.fft2(img)
```

```python
%timeit fft2 = np.fft.fft2(img, [nrows, ncols])
```

下面进行OpenCV的性能比较：

```python
%timeit dft1 = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
```

```python
%timeit dft2 = cv2.dft(np.float32(nimg), flags=cv2.DFT_COMPLEX_OUTPUT)
```

可以看到填充之后的DFT变换性能更佳。

# 5. 为什么Laplacian是高通滤波器
下面将不同类型的滤波器用傅里叶变换到图像，通过观察滤波器的频域图像可以进行分析：

```python
# 均值滤波器
mean_filter = np.ones((3, 3))

# 高斯滤波器
x = cv2.getGaussianKernel(5, 10)
gaussian = x * x.T

# 不同的边缘滤波器：
# x方向的Scharr滤波器
scharr = np.array([[-3, 0, 3],
                   [-10, 0, 10],
                   [-3, 0, 3]])

# x方向Sobel滤波器
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# y方向Sobel滤波器
sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

# laplacian
laplacian = np.array([[0, 1, 0],
                      [1, -4, 0],
                      [0, 1, 0]])

filters = [mean_filter, gaussian, scharr, sobel_x, sobel_y, laplacian]
filter_name = ['mean', 'gaussian', 'scharr', 'sobel_x', 'sobel_y', 'laplacian']

fft_filters = [np.fft.fft2(x) for x in filters]
fft_shift = [np.fft.fftshift(y) for y in fft_filters]
mag_spectrum = [np.log(np.abs(z) + 1) for z in fft_shift]

for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(mag_spectrum[i], cmap='gray')
    plt.title(filter_name[i]), plt.xticks([]), plt.yticks([])
```

从图像中，可以看到每个核的频率区域。通过这个信息可以看到每个核是高通滤波器还是低通滤波器。
