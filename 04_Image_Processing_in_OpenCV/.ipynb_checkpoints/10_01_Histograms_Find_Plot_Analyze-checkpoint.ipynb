````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
````

# 目标
- 分别使用Numpy和OpenCV计算直方图。
- 用OpenCV和Matplotlib绘图。
- 涉及函数：**cv2.calcHist()**和**cv2.histogram()**。

# 1. 理论
图像的直方图表示了图像中全部像素的亮度分布。它是以像素值（0-255）为X轴，对应像素数量作为Y轴。

这是另外一种理解图像的方式，通过图像直方图，可以获得对比度、亮度、亮度分布等图像信息。几乎所有图像处理工具都提供了直方图特征。下面的图像来自[Cambridge in Color website](http://www.cambridgeincolour.com/tutorials/histograms1.htm)。![](./imgs/histogram_sample.jpg)

该直方图是通过灰度图提取的，直方图左侧表示包含大量深色的像素，右边的区域表示了较明亮的像素。从中可以看到暗色区域要大于亮色区域。而中间的像素（像素值为127左右）数量非常少。

# 2. 计算直方图
前面介绍了是什么是直方图，接下来介绍如何计算直方图。OpenCV和Numpy都有计算直方图的内置函数。在使用这些函数之间，首先了解一下相关术语：

**BINS：** 上面的直方图显示了所有像素值对应的像素数量（从0到255），也就是说需要256个值来显示上面的直方图。但是有些情况下，我们不想计算所有像素值对应的像素数量，而是需要计算像素值区间所包含的像素数量。例如想要计算0-15、16-30、……、240-255各个区间的像素数量。此时只需要16个值来表示直方图。

在这种情况下，要做的就是把整个像素值区间分为16个子区间，然后计算各个子区间像素的数量。每个子区间称为一个**BIN**。在第一个例子中需要256个**BINS**（每个像素一个BIN），第二个例子中需要16个**BINS**。在OpenCV中用**histSize**来表示**BINS**的数量。

**DIMS：**表示要收集的参数数量，在上面的例子中，我们只收集亮度值的信息，因此设置为1。

**RANGE：**表示想要统计的亮度值范围，通常如果考虑所有亮度值，该值设置为$[0, 256]$。

## 2.1 用OpenCV计算直方图
函数**cv2.calcHist()**用于计算直方图，首先看一下它的函数定义：
$$cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])$$
1. **image** - 类型为**uint8**或**float32**的原始图像，需要用方括号括起来：**[img]**。
2. **channels** - 也要放到方括号中，是需要计算直方图的通道索引。例如，对于灰度图像它的值设置为**[0]**，对于彩色图像可以分别把**[0]**、**[1]**、**[2]**分别传递给函数来计算BGR三个通道的直方图。
3. **mask** - 掩码图像。要计算整幅图像的直方图，将其设置为None即可。但是如果要计算特定区域的直方图，需要创建一个掩码图像并传递给该函数。
4. **histSize** - 表示了**BIN**的数量，需要用方括号括起来，例如**[256]**。
5. **ranges** - 就是前面提到的**RANGE**，通常设置为**[0, 256]**

下面的代码加载了一幅灰度图像，并计算其直方图：

````python
img = cv2.imread('../data/home.jpg', cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist.shape
````

hist是一个256x1的数组，每个值对应着相应像素值的像素数量。

## 2.2 用Numpy计算直方图
Numpy提供了**np.histogram()**函数来计算直方图。

````python
hist, bins = np.histogram(img.ravel(), 256, [0, 256])
print(hist.shape)
print(bins.shape)
````

**hist**与OpenCV中的**cv2.calcHist()**函数计算的结果相同，但是**bins**有257个元素，这是因为numpy计算的**BINS**为：0-0.99，1-1.99，……，255-255.99。最后一个range为255-255.99，为了表达这一数据，numpy为bins添加了256，但是我们只需要到255即可。

**其他函数**
Numpy提供了另一个函数**np.bincount()**，比**np.histogram()**要快10倍左右。对于一维直方图，最好用这个函数。在调用该函数时**minlength=256**。例如：
``` Python
hist = np.bincount(img.ravel(), minlength=256)
```

> **注意**  
> OpenCV的函数比**np.histogram**要快40倍左右，因此需要坚持使用OpenCV函数。

# 3. 绘制直方图
有两种方式来绘制直方图：
1. 简单的绘制方式：用Matplotlib的绘制函数；
2. 复杂的绘制方式：用OpenCV绘制函数。

## 3.1 使用matplotlib
matplotlib提供了一个直方图绘制函数：**matplotlib.pyplot.hist()**。

如果要直接计算并绘制直方图，不需要调用**cv2.calcHist**或**np.histogram**函数，代码如下：

````python
img = cv2.imread('../data/home.jpg', cv2.IMREAD_GRAYSCALE)

plt.subplot(121)
plt.imshow(img, 'gray')
plt.subplot(122)
_ = plt.hist(img.ravel(), 256, [0, 256])
````

也可以用标准的matplotlib绘图方式，适用于绘制BGR不同通道的直方图。为了完成这一目的，首先计算直方图数据：

````python
img = cv2.imread('../data/home.jpg')
color = ['b', 'g', 'r']
for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
````

## 3.2 使用OpenCV
可以把**bin**值和**histogram**的值看做图像的坐标，然后用**cv2.line()**或**cv2.polyline()**来进行绘图。可以参考OpenCV源码中的Samples：[hist.py](./ref/hist.py)

# 4. 应用掩码图像
使用**cv2.calcHist()**可以计算整幅图像的直方图。如果想要计算某个区域的直方图，可以创建一幅掩码图像，白色的区域是需要计算直方图的区域，而黑色则不参与计算，然后将掩码图像传递给**cv2.calcHist()**函数。

````python
img = cv2.imread('../data/home.jpg', cv2.IMREAD_GRAYSCALE)

# 创建掩码图像
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300, 100:400] = 255
masked_img = cv2.bitwise_and(img, img, mask=mask)

# 分别计算有掩码和无掩码的直方图
# 注意第三个参数
hist_full = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_mask = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask, 'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0, 256])
````
