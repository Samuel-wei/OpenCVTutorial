# 1. 变换
OpenCV提供了两个变换函数：**cv2.warpAffine**和**cv2.warpPerspective**，可以做任何的变换。**cv2.warpAffine**使用一个2x3的变换矩阵作为输入，而**cv2.warpPerspective**使用一个3
3的变换矩阵作为输入。

## 1.1 缩放
缩放只是重新调整图像的尺寸。OpenCV使用**cv2.resize()**函数来实现这一目的。图像的尺寸可以手动的分别指定，或者指定一个缩放因子。缩放中可以使用不同的插值方法。缩小图像建议使用**cv2.INTER_AREA**，放大图像推荐使用**cv2.INTER_CUBIC**（较慢）或**cv2.INTER_LINEAR**。默认情况下，对于改变图像尺寸大小使用**cv2.INTER_LINEAR**。可以使用下面两种方法来改变图像的尺寸：

```python
import cv2
import numpy as np

img = cv2.imread('../data/messi5.jpg')
plt.imshow(img[:, :, ::-1])
```

```python
res = cv2.resize(img, None, fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
plt.imshow(res[:, :, ::-1])
```

```python
height, width = img.shape[:2]
res = cv2.resize(img, (2 * width, 2 * height), interpolation=cv2.INTER_CUBIC)

plt.imshow(res[:, :, ::-1])
```

## 1.2 平移
平移是用来移动目标的位置。如果已知平移的方向为$(x,y)$，用该方向作为仿射变换矩阵的$(t_x, t_y)$，可以构建变换矩阵$M$为：
$$M = \begin{bmatrix}
1 & 0 & t_x \\
0 & 1 & t_y
\end{bmatrix}$$

可以用一个类型为**np.float32**的Numpy数组作为变换矩阵传递给**cv2.warpAffine()**函数，下面的例子中图像的平移方向为$(100, 50)$：

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)
rows, cols = img.shape

M = np.float32([[1, 0, 100], [0, 1, 50]])
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst, cmap='gray')
```

**警告**  
**cv2.warpAffine()**函数的第三个参数用来指定输出图像的尺寸，注意它的格式为$(width, height)$。其中width是行数，height是列数。

## 1.3 旋转
将图像旋转$\theta$度是由下面的矩阵实现的：
$$M = \begin{bmatrix}
cos\theta & -sin\theta \\
sin\theta & cos\theta
\end{bmatrix}$$

但是OpenCV提供了绕某一像素进行缩放旋转的功能，需要将上面的旋转矩阵修改如下：
$$\begin{bmatrix}
\alpha & \beta & (1 - \alpha) \cdot center.x - \beta \cdot center.y \\
-\beta & \alpha & \beta \cdot center.x + (1 - \alpha) \cdot center.y
\end{bmatrix}$$
其中：
$$\begin{matrix}
\alpha = scale \cdot cos\theta \\
\beta = scale \cdot sin\theta
\end{matrix}$$

为了获得变换矩阵，OpenCV提供了**cv2.getRotationMatrix2D()**函数，下面的例子将图像以图像中心为原点旋转90度，无缩放：

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)
rows, cols = img.shape

M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst, cmap='gray')
```

## 1.4 仿射变换
在仿射变换中，原始图像中所有的平行线在输出图像中继续保持平行。要找到该变换矩阵，需要提供输入图像的三个点的位置以及在输出图像中的对应位置。然后用**cv2.getAffineTransform()**来创建一个2x3的矩阵，然后将其传递给**cv2.warpAffine()**函数。

下面的例子：

```python
img = cv2.imread('../data/drawing.png')
rows, cols, ch = img.shape

# 定义两组点
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

# 在原始图像上绘制上三个点
cv2.circle(img, (50, 50), 10, (0, 255, 0), -1, cv2.LINE_AA)
cv2.circle(img, (200, 50), 10, (0, 255, 0), -1, cv2.LINE_AA)
cv2.circle(img, (50, 200), 10, (0, 255, 0), -1, cv2.LINE_AA)
plt.imshow(img[:, :, ::-1])
```

```python
M = cv2.getAffineTransform(pts1, pts2)
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst[:, :, ::-1])
```

## 1.5 射影变换
对于射影变换，需要一个3x3的矩阵。在摄影变换中直线依旧保持为直线（平行关系不再保持）。为了得到射影变换矩阵，需要输入图像中的四个点和输出图像中对应的四个点。这四个点要保持任意三点不共线。可以用**cv2.getPerspectiveTransform()**函数来计算射影变换矩阵。然后将这个3x3矩阵传递给**cv2.warpPerspective()**函数。

```python
img = cv2.imread('../data/drawing.png')
rows, cols, ch = img.shape

pts1 = np.float32([[50, 100], [300, 200], [100, 900], [600, 700]])
pts2 = np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])

# 在原图像上绘制四个点
cv2.circle(img, (50, 100), 20, (0, 255, 0), -1, cv2.LINE_AA)
cv2.circle(img, (300, 200), 20, (0, 255, 0), -1, cv2.LINE_AA)
cv2.circle(img, (100, 900), 20, (0, 255, 0), -1, cv2.LINE_AA)
cv2.circle(img, (600, 700), 20, (0, 255, 0), -1, cv2.LINE_AA)
plt.imshow(img[:, :, ::-1])
```

```python
M = cv2.getPerspectiveTransform(pts1, pts2)
dst = cv2.warpPerspective(img, M, (300, 300))

plt.imshow(dst[:, :, ::-1])
```
