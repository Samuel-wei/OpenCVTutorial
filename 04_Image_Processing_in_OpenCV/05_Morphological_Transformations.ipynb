```python
from __future__ import print_function
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

# 目标
- 学习不同的形态学操作，例如腐蚀、膨胀、开运算、闭运算等等。
- 涉及以下函数：**cv2.erode()**、**cv2.dilate()**、**cv2.morphologyEx()**等等。

# 1. 理论
形态学变换是基于图像形状的简单操作。通常用于二值图像的操作。形态学变换需要两个输入，一个是源图像，另一个是决定操作性质的**结构元素**或**核**。两种最基本的形态学操作是**腐蚀**和**膨胀**。还包括开运算、闭运算和梯度等。下面将介绍这些操作。

## 1.1 腐蚀
腐蚀类似于土壤侵蚀的概念，该操作腐蚀前景对象（前景为白色）的边界。如何2D卷积一样，核滑过图像，只有当卷积核覆盖区域都是1时，对应像素的值才为1，否则被腐蚀掉（置为0）。

因此边界附近的像素将被腐蚀，前景对象将会变小，或者简单地说图像中的白色区域将会变小。该运算对于去除小的白噪声、分离两个连在一起的对象非常有效。

下面的例子中，将用5x5的全为1的核进行腐蚀操作：

```python
img = cv2.imread('../data/mor.png', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), dtype=np.uint8)
erosion = cv2.erode(img, kernel, iterations=1)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(erosion, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

## 1.2 膨胀
膨胀操作和腐蚀操作恰好相反，只要像素邻域内至少一个像素为1，该像素的值就为1。因此该操作会增加图像中白色区域的面积，或前景物体的大小。通常在噪声去除中，先做腐蚀再做膨胀。因为腐蚀会去除白噪声并缩小前景物体，然后在膨胀前景物体。由于噪声已经被消除，因此膨胀时噪声不会再出现，而前景物体得到了增长。膨胀操作对于连接被断开的物体非常有用。

```python
img = cv2.imread('../data/mor.png', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), dtype=np.uint8)
dialation = cv2.dilate(img, kernel, iterations=1)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(dialation, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

## 1.3 开运算
开运算的本质就是：**腐蚀之后膨胀**。用于去除噪声，可以用**cv2.morphologyEx()**函数来执行该操作。

```python
img = cv2.imread('../data/mor_opening.jpg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), dtype=np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(opening, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

## 1.4 闭运算
闭运算是开运算的逆运算，先做膨胀在进行腐蚀。该运算用于去除前景上的小洞，或者目标上的黑点。

```python
img = cv2.imread('../data/mor_closing.jpg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), dtype=np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(closing, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

## 1.5 形态学梯度
该运算计算一幅图像膨胀和腐蚀操作后的差异，下面的例子展示了对图像应用形态学梯度的效果。

```python
img = cv2.imread('../data/mor.png', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), dtype=np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(gradient, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

## 1.6 Top Hat
该运算计算输入图像和图像开运算的差异，下面的例子采用了9x9的核。

```python
img = cv2.imread('../data/mor.png', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((9, 9), dtype=np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(tophat, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

# 1.7 Black Hat
该操作计算了图像闭运算和输入图像之间的差异：

```python
img = cv2.imread('../data/mor.png', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((9, 9), dtype=np.uint8)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

plt.subplot(121), plt.imshow(img, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Origin Image')

plt.subplot(122), plt.imshow(blackhat, 'gray')
plt.xticks([]), plt.yticks([]), plt.title('Erosion Image')
```

# 2. 结构元素
之前的例子中用Numpy创建了结构元素，该结构元素是矩形的。但在某些情况，可能需要椭圆或圆形的核。可以用**cv2.getStructuringElement()**来根据核的尺寸生成需要的核。

```python
cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
```

```python
cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
```

```python
cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
```
