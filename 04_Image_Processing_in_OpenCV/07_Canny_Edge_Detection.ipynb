# 目标
- Canny边缘检测的概念。
- 涉及的OpenCV函数：**cv2.Canny()**。

# 1. 理论
Canny边缘检测是由John F. Canny提出的一个非常流行的边缘检测算法。
1. 该算法由多个步骤构成，下面介绍每一个步骤。  
2. **降噪：**由于边缘检测易于受到图像噪声的影响，因此首先用$(5, 5)$的卷积核进行高斯滤波。  
3. **计算图像亮度的梯度：**对平滑后的图像用**Sobel**滤波器来计算水平和垂直方向的一阶梯度，水平方向的一阶梯度记为$G_x$，垂直方向的一阶梯度记为$G_y$。用这两幅图像，可以计算出每个像素的边缘梯度和方向：
$$\text{Edge_Gradient}(G)=\sqrt {G_x^2 + G_y^2}$$
$$\text{Angle}(\theta) = tan^{-1}( \frac {G_y} {G_x})$$
梯度方向总是与边缘垂直。然后四舍五入为四个方向：水平、垂直和两个对角线方向。
4. **非最大值抑制：**在获得梯度的强度和方向后，对图像进行遍历去除不需要的像素（没有用于构成边缘的像素）。因此，在每一个像素，检查该像素在梯度方向的邻域上是否是局部最大值。检查方法如下图所示：![](./imgs/canny_01.jpg)
在边缘（垂直的边）上的点A，其梯度方向与边缘垂直。点B和点C都在梯度方向上，因此点A要检查点B和点C的强度来确定自己是否是局部最大值。如果是保留该点用于下一步的处理，否则进行抑制（置为0）。
简单地说，这一部分的结果是具有**thin edges**的二值图像。  
5. **迟滞阈值：**这一步骤的目的是确定所有的边缘点是否是真的边缘。需要两个阈值**minVal**和**maxVal**来完成这一功能。任何亮度梯度大于**maxVal**的边缘点被确定为**Sure-Edge**，而亮度梯度小于**minVal**的边缘点则被丢弃。位于两个阈值之间的边缘点在确实它们是否真的属于边缘时，要考虑它们的连通性。如果该点与**Sure-Edge**像素连通，则被作为边缘的一部分，而没有连通到**Sure-Edge**的点则被弃用。如下图所示：![](./imgs/canny_02.jpg)点A大于**maxVal**，因此被指定为**Sure-Edge**。尽管点C小于**maxVal**，但是与点A连通，因此被认为是边缘的一部分，这样我们就可以得到一个完整的边缘曲线。而点B尽管大于**minVal**，但是没有与**Sure-Edge**连通，因此被丢弃。因此为了获得正确的结果，必须选择合适的**maxVal**和**minVal**。
由于假设边缘是较长的线，因此移除小的像素噪点。

最后，就获得了图像的强边缘。

# 2. OpenCV中的Canny边缘检测
OpenCV将上述步骤封装到一个函数中：**cv2.Canny()**。该函数第一个参数是输入图像，第二个和第三个参数是**minVal**和**maxVal**。第四个参数**aperture_size**表示Sobel核的大小，默认设置为3。最后一个参数**L2gradient**指定计算亮度梯度的公式。如果设置为True，使用上面提到的公式，该公式具有较高的精度。否则使用函数：$\text{Edge_Gradient}(G) = |G_x|+|G_y|$，默认值为**False**。

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(img, 100, 200)

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Canny Edge'), plt.xticks([]), plt.yticks([])
```
