# 目标
- 读取图像、显示图像和存储图像
- 涉及的函数：**cv2.imread()**、**cv2.imshow()**和**cv2.imwrite()**
- 可选内容：用Matplotlib显示图像

# 1. 使用OpenCV
## 1.1 读取图像
可以使用**cv2.imread()**函数来读取图像。图像需要位于当前的工作目录，或指定完整的图像的路径。

该函数的第二个参数读取图像的方式：
- **cv2.IMREAD_COLOR**：加载彩色图像，图像的透明度将会被忽略，是默认参数。
- **cv2.IMREAD_GRAYSCALE**：以灰度的形式加载图像。
- **cv2.IMREAD_UNCHANGED**：加载图像是包含alpha通道（透明度通道）

> **注意**  
> 除了使用上述三个标志，也可以用分别用1、0和-1来代替他们。

参阅以下代码：

```python
# 加载必要的库
import numpy as np
import cv2
```

```python
# 将一张彩色图像加载为灰度图像
img = cv2.imread('../data/messi5.jpg', 0)
print(img.shape)
```

**警告**  
即使图像的路径错误，也不会报错。但是打印**img**结果为：**None**。

## 1.2 显示图像
可以使用**cv2.imshow()**在一个窗口中显示图像，窗口自动调整大小以适应图片。

该函数的第一个参数是字符串类型的窗口名字。第二个参数是我们的图像。创建窗口的数量没有限制，但是窗口需要不同的名字。

```python
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

窗口的截图如下图所示（在运行OS-X-Yosemite机器）：
![](imgs/01_1.jpg)

**cv2.waitKey()**是一个键盘绑定函数。它的参数是以毫秒为单位的时间。该函数等待指定毫秒数键盘事件。如果在那个事件按下了任意键，程序将继续。如果设置为0，将会产生键盘阻塞。该函数也可设置为对指定按键阻塞。

> **注意**  
> 除了键盘绑定，该函数还处理很多其他的GUI时间，因此**必须**使用它来显示图像。

**cv2.destroyAllWindows()**将我们创建的所有窗口销毁。如果想要销毁特定的窗口，可以将要摧毁窗口的名字作为参数传递给**cv2.destroyWindow()**。

> **注意**  
> 还有一种特殊的情况是你已经创建了窗口，之后要将图像显示在该窗口中。在这种情况下，你可以指定窗口是否可以改变大小。可以使用**cv2.namedWindow()**实现这一功能。默认情况下，标志被设置为**cv2.WINDOW_AUTOSIZE**。但是可以将其指定为**cv2.WINDOW_NORMAL**，就可以改变窗口的大小了。这对于显示较大的图片或为窗口添加拖动条时非常有用。

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 1.3 写入图像
可以使用**cv2.imwrite()**来写入图像。

第一个参数是文件名，第二个参数是需要存储的图像。

```python
cv2.imwrite('messigray.png', img)
```

上面的代码将图像存储为png格式。

## 1.4 集成
下面的程序以灰度的格式读取图像，然后显示。如果按下*s*键则存储图像并退出，若按下*ESC*则直接退出。

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('image', img)
k = cv2.waitKey(0) & 0xFF
if k == 27: # ESC
    cv2.destroyAllWindows()
elif k == ord('s'): # s
    cv2.imwrite('messigray.png', img)
    cv2.destroyAllWindows()
```

**警告**
在64位的机器上，需要**cv2.waitKey(0) & 0xFF**，而32位的机器只需要**cv2.waitKey(0)**即可。

# 2. 使用Matplotlib
Matplotlib是Python中的绘图库，该库提供了大量的绘图方法。下面学习如何用Matplotlib显示图像。

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

img = cv2.imread('../data/messi5.jpg', 0)
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

**警告**  
OpenCV读取的图像是**BGR**模式，而Matplotlib显示**RGB**模式。因此彩色图像无法再其中正确显示。

```python
img = cv2.imread('../data/messi5.jpg', cv2.IMREAD_COLOR)
plt.imshow(img)
```

因此需要改变其通道：

```python
b, g, r = cv2.split(img)
img_rgb = cv2.merge([r, g, b])
plt.imshow(img_rgb)
```

或者直接用numpy的操作来快速转换:

```python
img_rgb = img[:, :, ::-1]
plt.imshow(img_rgb)
```
