```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# 定义一个用于在matplotlib中显示图像的函数
def imshow(img):
    plt.imshow(img[:, :, ::-1])
```

# 目标
- 学习一些图像的算数运算，例如加法、减法和按位运算等。
- 涉及的函数：**cv2.add()**、**cv2.addWeighted()**等。

# 1. 图像的加法
可以用OpenCV中的函数**cv2.add()**或者直接使用Numpy的操作：**res = img1 + img2**。参与加法运算的图像需要有相同的通道数和类型，或者第二个图像是一个标量。

> **注意**  
> OpenCV的加法和Numpy的加法之间的差别是：OpenCV的加法是饱和运算，而Numpy的加法是模运算。

看下面的例子：

```python
x = np.uint8([250])
y = np.uint8([10])

print(cv2.add(x, y)) # 250 + 10 = 260 => 255
print(x + y)         # 250 + 10 = 260 % 256 = 4
```

从上面的例子可以看出，OpenCV的函数能够提供更好的结果，因此要坚持使用OpenCV函数。

# 2. 图像混合
图像混合本质上也是图像的加法，但是可以为每个图像指定权重，这样可以得到混合或半透明的效果。图像采用下面的公式进行求和：
$$g(x)=(1 - \alpha)f_0(x) + \alpha f_1(x)$$
其中$\alpha \in [0, 1]$，可以进行从一张图像过渡到另一张图像的操作。

下面将两张图像混合到一起，第一个图像的权重为$0.7$，第二张图像的权重为$0.3$。**cv2.addWeighted()**应用下面的公式：
$$dst = \alpha \cdot img1 + \beta \cdot img2 + \gamma$$
下面的例子中$\gamma$设置为$0$：

```python
img1 = cv2.imread('../data/ml.png')
img2 = cv2.imread('../data/opencv_logo.jpg')



plt.subplot(121), imshow(img1)
plt.subplot(122), imshow(img2)
```

```python
dst = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
imshow(dst)
```

# 3. 按位运算
按位运算包含按位的**AND**、**OR**、**NOT**和**XOR**。在提取图像的某个部分的操作时非常有用，尤其是要定义不规则的ROI时。下面的例子将展示如何改变图像的特定区域。

```python
img1 = cv2.imread('../data/messi5.jpg')
imshow(img1)
```

```python
img2 = cv2.imread('../data/opencv-logo-white.png')
imshow(img2)
```

```python
# 要将图像放置到图像的左上角，因此创建一个ROI
rows, cols, channels = img2.shape
roi = img1[0:rows, 0:cols]

# 创建一个logo的mask，然后创建反mask
img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)

# 把黑色的区域从ROI中去掉
img1_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)

# 把logo从logo图像中提取出来
img2_fg = cv2.bitwise_and(img2, img2, mask=mask)

# 将logo放入roi，然后修改原图像
dst = cv2.add(img1_bg, img2_fg)
img1[0:rows, 0:cols] = dst

imshow(img1)
```

```python
plt.subplot(121), imshow(img1_bg), plt.title('img bg')
plt.subplot(122), imshow(img2_fg), plt.title('logo fg')
```
